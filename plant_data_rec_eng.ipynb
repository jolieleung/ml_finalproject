This dataset was previously cleaned in OpenRefine to split columns that express values in a range into separate min/max columns. Here we should drop those original columns since they are still expressed as objects instead of floats.

```python
plant_data = plant_data.drop(columns=['height (feet)', 'spread (feet)', 'bloom_time', 'sun', 'zone'])
```

Convert Bloom Start and Bloom End to Date-Time

```python
#converted strings into numbers, and thent ried to convert into datetime. these are still showing up as objects though?
months = {'January' : 1, 'February' : 2, 'March' : 3, 'April' : 4, 'May' : 5, 'June' : 6, 'July' : 7, 'August' : 8, 'September' : 9, 'October' : 10, 'November' : 11, 'December' : 12}

bloom_clean = {"bloom_start" : months,
              "bloom_end" : months}
```

```python
plant_data = plant_data.replace(bloom_clean)
```
