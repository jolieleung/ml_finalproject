## 1. Read in & clean data

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plant_data = pd.read_csv("all-plant-data.csv")
plant_data
```

This dataset was previously cleaned in OpenRefine to split columns that express values in a range into separate min/max columns. Here we should drop those original columns since they are still expressed as objects instead of floats.

```python
plant_data = plant_data.drop(columns=['height (feet)', 'spread (feet)', 'bloom_time', 'sun', 'zone','native_range'])
```

```python
plant_data.dtypes
```

```python
plant_data.isna().sum()
```

Convert Bloom Start and Bloom End to Date-Time

```python
plant_data["bloom_start"] = plant_data["bloom_start"].astype(str)
plant_data["bloom_end"] = plant_data["bloom_end"].astype(str)
```

```python
plant_data[["bloom_start", "bloom_end"]] = plant_data[["bloom_start", "bloom_end"]].apply(pd.to_datetime, format="%B", errors="coerce")
```

```python
# extracting month

plant_data["bloom_start_month"] = plant_data["bloom_start"].apply(lambda x: x.month)
plant_data["bloom_end_month"] = plant_data["bloom_end"].apply(lambda x: x.month)
```

```python
plant_data
```

```python
plant_data = plant_data.drop(columns=['bloom_start', 'bloom_end'])
```

```python
plant_data = plant_data.set_index("scientific_name")
```

```python
plant_data
```

```python
plant_data.dtypes
```

```python
plant_data_nona = plant_data.dropna()
```

Encode categorical features and normalizing data.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numeric_features = plant_data.select_dtypes(exclude='O').columns

numeric_transformer = Pipeline(
    steps=[("scaler", StandardScaler())]
)

categorical_features = plant_data.select_dtypes(include='O').columns

categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)
```

```python
processing_pipe = Pipeline(
    steps=[("preprocessor", preprocessor)]
)
```

```python
processed_plant_data = processing_pipe.fit_transform(plant_data)
```

Convert data to an array and reassign index to processed dataframe.

```python
plant_data_df = pd.DataFrame(processed_plant_data.toarray())
plant_data_df.index = plant_data.index
```

```python
plant_data_df
```

```python
plant_data_df.isna().sum()
```

```python
plant_data_df[plant_data_df.notnull()]
```

```python
plant_data_df_nona = plant_data_df.dropna()
```

```python
plant_data_df_nona
```

## 2. Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_matrix = cosine_similarity(plant_data_df_nona, plant_data_df_nona)
```

```python
sim_matrix.shape
```

```python
df_sim = pd.DataFrame(sim_matrix, columns=plant_data_df_nona.index, index=plant_data_df_nona.index)
```

```python
df_sim.head()
```

```python
def recommend(plant):
    plant_id=plant_data_df_nona.index
    scores=list(enumerate(df_sim[plant_id]))
    sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
    sorted_scores=sorted_scores[1:]
    plants=[plant_data_df_nona.index for plants in sorted_scores]
    return plants
```

```python
def recommend_five(plant_list):
    first_five=[]
    count=0
    for plant in plant_list:
        if count > 4:
            break
        count+=1
        first_five.append(plant)
    return first_five
```
